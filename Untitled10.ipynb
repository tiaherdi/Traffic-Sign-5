```python
#Grouping the dataset
Grouped=df.groupby(df.Country)
Can=Grouped.get_group("Canada")
Can1=pd.DataFrame(Can)
```

```python
Can_timeseries= Can1[["Date","PercentOfBaseline"]]
Can_timeseries['Date']=pd.to_datetime(Can_timeseries['Date'])#preparing data
Can_timeseries.set_index('Date',inplace=True)
```

```python
! pip install pmdarima
import pmdarima as pm
from pmdarima.arima import ADFTest
ADF=ADFTest(alpha=0.05)
ADF.should_diff(Can_timeseries) #non staionary data
```

```python
#TrainTest split
train=Can_timeseries[:1848]
test=Can_timeseries[1849:]
#Building Arima
from pmdarima import auto_arima
Can_Arima= auto_arima(train,start_p=0,d=1,start_q=0,max_p=5,max_d=5,max_q=5,start_P=0,D=1,start_Q=0,max_P=5,max_D=5,max_Q=5,m=12,seasonal=False,error_action='warn',
             trace=True,suppress_warnings=True,stepwise=True,random_state=55,n_fits=50 )
```

```python
Can_Arima.summary()
```

```python
Can_forecast=pd.DataFrame(Can_Arima.predict(n_periods=462),index=test.index)
Can_forecast.columns=['Forecasted']
```

```python

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import math
print(mean_absolute_error(test['PercentOfBaseline'],Can_forecast))
print(math.sqrt(mean_squared_error(test['PercentOfBaseline'],Can_forecast)))
```
