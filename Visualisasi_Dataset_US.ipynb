```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/covid_impact_on_airport_traffic.csv')
df.head()
pd.to_datetime(df['Date'])
```

```python
# based on country
# data (date, percent of baseline, country)

# style
sns.set(font_scale=1)
sns.set_style("whitegrid")

# plot
fig, ax = plt.subplots(figsize=(20,3))
by_country = df.query("Country == 'United States of America (the)'")
sns.lineplot(data=by_country, x='Date', y='PercentOfBaseline')
```

```python
# based on airport Los Angeles International
# data (date, percent of baseline, country)

# style
sns.set(font_scale=1)
sns.set_style("whitegrid")

# plot
fig, ax = plt.subplots(figsize=(10,3))
by_airport = df.query("AirportName == 'Los Angeles International'")
sns.lineplot(data=by_airport, x='Date', y='PercentOfBaseline', hue='AirportName')
```
